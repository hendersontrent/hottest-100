```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import gaussian_process
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, WhiteKernel, ConstantKernel
```

```python
# Load data

d = pd.read_excel("/Users/trenthenderson/Documents/Git/hottest-100/data/full_data_set.xlsx")

# Filter for each nationality

aus = d[d['australian'] == 1]
inter = d[d['australian'] == 0]

# Log-scale variables and make numpy arrays for modelling

aus = aus[['plays', 'fb_likes']]
inter = inter[['plays', 'fb_likes']]

aus['plays'] = np.log(aus['plays'])
aus['fb_likes'] = np.log(aus['fb_likes'])

inter['plays'] = np.log(inter['plays'])
inter['fb_likes'] = np.log(inter['fb_likes'])

aus = aus.to_numpy()
inter = inter.to_numpy()

# Reshape for modelling

aus_x = aus[:,1]
aus_x = aus_x.reshape(-1, 1)
aus_y = aus[:,0]

int_x = inter[:,1]
int_x = int_x.reshape(-1, 1)
int_y = inter[:,0]
```

```python
# International artist model

kernel = ConstantKernel() + Matern(length_scale=2, nu=3/2) + WhiteKernel(noise_level=1)

clf = GaussianProcessRegressor(random_state = 123, kernel = kernel)
clf.fit(int_x, int_y)
y_pred = clf.predict(int_x)

fig, ax = plt.subplots(1, figsize=(12, 6))
plt.plot(int_x[:, 0], int_y, marker = 'o', color = 'black', linewidth = 0)
plt.plot(int_x[:, 0], y_pred, marker = 'x', color = 'steelblue')
plt.suptitle("$GaussianProcessRegressor(kernel=RBF)$ [default]", fontsize = 20)
plt.axis('off')
```
